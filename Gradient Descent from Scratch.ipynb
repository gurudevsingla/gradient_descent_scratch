```python
import math
import random
import numpy as np
```

```python
def f(x):
    return math.pow(x,2)+2*x+10
```

```python
def derivative(x):
    return 2*x+2
```

```python
def gradient_descent(eta):
    
    x0 = random.random()
    i = 0
    
    xi = x0
    df_dx = derivative(xi)
    
    while abs(df_dx-0.0)>0.001:
        print(i)
        print(xi)
        print(df_dx)
        print("*"*25)
        
        df_dx = derivative(xi)
        xi = xi - eta*df_dx
        i+=1
        
    return round(xi,2)
```

```python
gradient_descent(0.1)
```
